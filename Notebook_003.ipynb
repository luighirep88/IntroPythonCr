```python
# Recordemos las librerias a usar

import numpy as np
import matplotlib.pyplot as plt

import pandas as pd # permite manejar matrices y algunas operaciones numericas especializados
```

```python

# level 1: Dispersión + línea de regresión lineal =====

# Para que los resultados sean reproducibles
np.random.seed(42)

# 1. Generamos datos aleatorios
n = 50  # número de observaciones

# "Horas de estudio": variable explicativa (X)
horas_estudio = np.random.uniform(0, 10, n)  # entre 0 y 10 horas

# "Calificación": variable respuesta (Y)
# Relación (más o menos): nota = 2.5 * horas + 10 + ruido
ruido = np.random.normal(0, 3, n)  # ruido aleatorio con media 0 y desviación 3
calificacion = 2.5 * horas_estudio + 10 + ruido

# 2. Ajustamos una regresión lineal rápida con numpy.polyfit
# polyfit(X, Y, grado_del_polinomio)
coef_lineal = np.polyfit(horas_estudio, calificacion, 1)
                         #m o pendiente
                         #a       y = a + m x
pendiente, intercepto = coef_lineal

print(f"Pendiente estimada: {pendiente:.2f}")
print(f"Intercepto estimado: {intercepto:.2f}")

# 3. Creamos puntos "suaves" para dibujar la línea de tendencia
x_linea = np.linspace(horas_estudio.min(), horas_estudio.max(), 100)
y_linea = pendiente * x_linea + intercepto

# 4. Graficamos
plt.figure(figsize=(8, 5))

# Dispersión de los datos
plt.scatter(horas_estudio, calificacion, label="Datos simulados", alpha=0.8)

# Línea de regresión
plt.plot(
    x_linea,
    y_linea,
    label=f"Línea de regresión\n(y = {pendiente:.2f}x + {intercepto:.2f})"
)

# Etiquetas y detalles estéticos
plt.xlabel("Horas de estudio")
plt.ylabel("Calificación")
plt.title("Relación entre horas de estudio y calificación (datos simulados)")
plt.legend()
plt.grid(True)

plt.show()
```

```python
# Nivel 2: Dispersión + curva polinómica (grado 2)

np.random.seed(123)

n = 60

# X: "Años de experiencia"
anios_experiencia = np.random.uniform(0, 20, n)

# Y: "Productividad" con relación cuadrática + ruido
# product = 0.3*x^2 + 2*x + 5 + ruido
ruido = np.random.normal(0, 15, n)
productividad = 0.3 * anios_experiencia**2 + 2 * anios_experiencia + 5 + ruido

# Ajustamos un polinomio de grado 2
coef_poly = np.polyfit(anios_experiencia, productividad, 2)
a, b, c = coef_poly

print(f"Coeficientes polinomio grado 2:")
print(f"  a (x^2): {a:.3f}")      #a(x^2) + bx + c = Y
print(f"  b (x):   {b:.3f}")
print(f"  c:       {c:.3f}")

# Creamos una malla de X para dibujar la curva suave
x_curva = np.linspace(anios_experiencia.min(), anios_experiencia.max(), 200)
y_curva = a * x_curva**2 + b * x_curva + c

# Graficamos
plt.figure(figsize=(8, 5))

plt.scatter(anios_experiencia, productividad, label="Datos simulados", alpha=0.8)
plt.plot(
    x_curva,
    y_curva,
    label="Tendencia polinómica (grado 2)",
    linewidth=2
)

plt.xlabel("Años de experiencia")
plt.ylabel("Productividad (índice simulado)")
plt.title("Gráfico de dispersión con tendencia polinómica de grado 2")
plt.legend()
plt.grid(True)

plt.show()
```

```python
# NIVEL 3: if anidado para segmentar personas +

def asignar_segmento(fila):
    edad = fila["edad"]
    ingreso = fila["ingreso_mensual"]
    estudiante = fila["es_estudiante"]

    if estudiante:
        # Primer gran bloque: estudiantes
        if ingreso < 1_200_000:
            return "estudiante vulnerable"
        else:
            return "estudiante estable"
    else:
        # Segundo bloque: no estudiantes
        if edad < 25 and ingreso >= 1_200_000:
            return "joven trabajador con ingreso aceptable"
        else:
            return "adulto trabajador"

df_personas["segmento"] = df_personas.apply(asignar_segmento, axis=1)

df_personas
```

```python
#  elif para clasificar por edad , util si queremos tener una idea diferente a la de cohortes, aunque algunos puristas detestan eso d elas generacioens y lo del ciclo vital

def clasificar_edad(edad):
    if edad < 18:
        return "menor de edad"
    elif edad < 25:
        return "joven"
    elif edad < 30:
        return "adulto joven"
    else:
        return "adulto"

df_personas["categoria_edad"] = df_personas["edad"].apply(clasificar_edad)

df_personas[["id", "edad", "categoria_edad"]]
```

```python
df_personas["categoria_edad"].value_counts()
```

```python
# L 2: elif para tramos de impuesto

#Declaro función tasa de impuesto...
def tasa_impuesto(ingreso):
    if ingreso < 1_000_000:
        return 0.00
    elif ingreso < 2_000_000:
        return 0.05
    elif ingreso < 3_000_000:
        return 0.10
    else:
        return 0.15

df_personas["tasa_impuesto"] = df_personas["ingreso_mensual"].apply(tasa_impuesto)
df_personas["impuesto_mensual"] = (df_personas["ingreso_mensual"] * df_personas["tasa_impuesto"]).round(0)

df_personas[["id", "ingreso_mensual", "tasa_impuesto", "impuesto_mensual"]]
```

```python
df_personas.groupby("tasa_impuesto")[["ingreso_mensual", "impuesto_mensual"]].mean().round(0)
```

```python
# NIVEL 3: elif con múltiples condiciones

def asignar_tipo_apoyo(fila):
    edad = fila["edad"]
    ingreso = fila["ingreso_mensual"]
    estudiante = fila["es_estudiante"]

    if estudiante and ingreso < 1_200_000:
        return "beca completa"
    elif estudiante and ingreso < 2_000_000:
        return "beca parcial"
    elif (not estudiante) and edad < 25 and ingreso < 1_500_000:
        return "renta solidaria joven trabajador"
    elif ingreso >= 2_500_000:
        return "sin apoyo por altos ingresos"
    else:
        return "apoyo básico"

df_personas["tipo_apoyo"] = df_personas.apply(asignar_tipo_apoyo, axis=1)

df_personas[["id", "edad", "ingreso_mensual", "es_estudiante", "tipo_apoyo"]]
```

```python
#N 2: acumular estadísticas con for
suma_ingresos = 0
conteo_personas = 0
conteo_estudiantes = 0

for idx, fila in df_personas.iterrows():
    ingreso = fila["ingreso_mensual"]
    estudiante = fila["es_estudiante"]

    suma_ingresos += ingreso         # sumamos ingresos
    conteo_personas += 1             # contamos personas

    if estudiante:
        conteo_estudiantes += 1      # contamos estudiantes

# Calculamos el ingreso promedio
if conteo_personas > 0:
    ingreso_promedio = suma_ingresos / conteo_personas
else:
    ingreso_promedio = 0

print(f"Número de personas: {conteo_personas}")
print(f"Suma total de ingresos: {suma_ingresos:,} pesos")
print(f"Ingreso promedio: {ingreso_promedio:,.0f} pesos")
print(f"Número de estudiantes: {conteo_estudiantes}")
print(f"Proporción de estudiantes: {conteo_estudiantes / conteo_personas:.2f}")
```

```python
# NIVEL 3: resumen por grupo usando for + diccionarios

resumen_por_estudiante = {}  # dict vacío

for idx, fila in df_personas.iterrows():
    clave = fila["es_estudiante"]          # True o False
    ingreso = fila["ingreso_mensual"]

    # Si la clave aún no existe en el dict, la creamos
    if clave not in resumen_por_estudiante:
        resumen_por_estudiante[clave] = {
            "suma_ingresos": 0,
            "conteo": 0
        }

    # Actualizamos acumulados
    resumen_por_estudiante[clave]["suma_ingresos"] += ingreso
    resumen_por_estudiante[clave]["conteo"] += 1

# Calculamos el ingreso promedio por grupo
for clave, datos in resumen_por_estudiante.items():
    suma = datos["suma_ingresos"]
    n = datos["conteo"]
    promedio = suma / n if n > 0 else 0

    print(f"es_estudiante = {clave}")
    print(f"  Personas: {n}")
    print(f"  Suma de ingresos: {suma:,} pesos")
    print(f"  Ingreso promedio: {promedio:,.0f} pesos")
    print("-" * 40)
```

```python
# resumen por perfil_riesgo =====

resumen_riesgo = {}

for idx, fila in df_personas.iterrows():
    perfil = fila["perfil_riesgo"]          # "alto riesgo", "riesgo medio", "bajo riesgo"
    ingreso = fila["ingreso_mensual"]

    if perfil not in resumen_riesgo:
        resumen_riesgo[perfil] = {
            "suma_ingresos": 0,
            "conteo": 0
        }

    resumen_riesgo[perfil]["suma_ingresos"] += ingreso
    resumen_riesgo[perfil]["conteo"] += 1

for perfil, datos in resumen_riesgo.items():
    suma = datos["suma_ingresos"]
    n = datos["conteo"]
    promedio = suma / n if n > 0 else 0

    print(f"Perfil de riesgo: {perfil}")
    print(f"  Personas: {n}")
    print(f"  Ingreso total: {suma:,} pesos")
    print(f"  Ingreso promedio: {promedio:,.0f} pesos")
    print("-" * 40)
```

## Actividad 06 el uso del While

```python
# L 1: while simple (ahorro hasta una meta)

meta_ahorro = 5_000_000   # meta en pesos
ahorro_mensual = 400_000  # cuánto puede ahorrar cada mes
saldo = 0
meses = 0

while saldo < meta_ahorro:
    meses += 1
    saldo += ahorro_mensual
    print(f"Mes {meses}: saldo acumulado = {saldo:,} pesos")

print("\nResultado:")
print(f"Se alcanzó la meta de {meta_ahorro:,} pesos en {meses} meses.")
```

```python
# NIVEL 2: while con intereses y límite de meses

meta_ahorro = 10_000_000      # meta en pesos
aporte_mensual = 500_000      # depósito cada mes
tasa_interes_mensual = 0.01   # 1 % mensual
saldo = 0
mes = 0
max_meses = 60                # límite de seguridad (5 años)

while saldo < meta_ahorro and mes < max_meses:
    mes += 1
    # Primero el aporte
    saldo += aporte_mensual
    # Luego aplicamos el interés sobre el saldo
    saldo = saldo * (1 + tasa_interes_mensual)

    print(f"Mes {mes}: saldo = {saldo:,.0f} pesos")

print("\nResultado:")
if saldo >= meta_ahorro:
    print(f"Se alcanzó o superó la meta de {meta_ahorro:,} pesos en {mes} meses.")
else:
    print(f"No se alcanzó la meta en {max_meses} meses. Saldo final: {saldo:,.0f} pesos.")
```

```python


np.random.seed(123)  # para reproducibilidad

n_estudiantes = 9

# Fecha base (ej. 1 de marzo de 2025 a las 8:00 a.m.)
fecha_base = pd.to_datetime("2025-03-01 08:00")

# Generamos minutos aleatorios de envío dentro de una semana
offset_minutos = np.random.randint(0, 7 * 24 * 60, size=n_estudiantes)

fechas_envio = [
    fecha_base + pd.Timedelta(minutes=int(m))
    for m in offset_minutos
]

df_estudiantes = pd.DataFrame({
    "id_estudiante": range(1, n_estudiantes + 1),
    "fecha_envio": fechas_envio,
    "nombre": [f"Estudiante_{i}" for i in range(1, n_estudiantes + 1)],
    "doc_identidad_ok": np.random.choice([True, False], size=n_estudiantes),
    "acta_grado_ok":   np.random.choice([True, False], size=n_estudiantes),
    "sisben_ok":       np.random.choice([True, False], size=n_estudiantes),
    "formulario_ok":   np.random.choice([True, False], size=n_estudiantes)
})

# Ordenamos por fecha_envio (orden de llegada) y reindexamos
df_estudiantes = df_estudiantes.sort_values("fecha_envio").reset_index(drop=True)

df_estudiantes

```

```python
# NIVEL 3: while seleccionando hasta 3 estudiantes aprobados



df_estudiantes["estado_solicitud"] = "no_revisada"
df_estudiantes["seleccionado"] = False  # marcar quién queda entre los 3 seleccionados

indice = 0
aprobados = 0
n_filas = len(df_estudiantes)
max_aprobados = 3

while indice < n_filas and aprobados < max_aprobados:
    fila = df_estudiantes.iloc[indice]

    print(f"Revisando solicitud de {fila['nombre']} (id {fila['id_estudiante']})")
    print(f"  Fecha de envío: {fila['fecha_envio']}")

    # Verificamos todos los documentos
    tiene_doc = fila["doc_identidad_ok"]
    tiene_acta = fila["acta_grado_ok"]
    tiene_sisben = fila["sisben_ok"]
    tiene_formulario = fila["formulario_ok"]

    if tiene_doc and tiene_acta and tiene_sisben and tiene_formulario:
        estado = "aprobada"
        seleccionado = True
        aprobados += 1
        print("  -> Solicitud APROBADA: todos los documentos están completos.")
        print(f"  -> Esta persona entra dentro del cupo de {max_aprobados} seleccionados.")
    else:
        estado = "rechazada"
        seleccionado = False
        print("  -> Solicitud RECHAZADA: falta al menos un documento.")

    # Actualizamos el DataFrame
    df_estudiantes.loc[indice, "estado_solicitud"] = estado
    df_estudiantes.loc[indice, "seleccionado"] = seleccionado

    indice += 1  # pasamos al siguiente en orden de fecha_envio

    print("-" * 60)

print("\nRevisión finalizada.")
print(f"Total de estudiantes aprobados: {aprobados}")
```

```python
# Importamos la librería pandas, que permite crear y manipular DataFrames
import pandas as pd

```

```python
# Creamos una lista de diccionarios.
# Cada diccionario representa un país con sus variables: PIB, población y área.
datos_paises = [
    {"pais": "Colombia",     "PIB": 314_000, "poblacion": 52_000_000, "area": 1_142_000},
    {"pais": "Brasil",       "PIB": 1_920_000, "poblacion": 203_000_000, "area": 8_516_000},
    {"pais": "Argentina",    "PIB": 632_000, "poblacion": 46_000_000, "area": 2_780_000},
    {"pais": "Chile",        "PIB": 317_000, "poblacion": 19_500_000, "area": 756_000}
]

# Convertimos la lista de diccionarios en un DataFrame
df = pd.DataFrame(datos_paises)

# Mostramos el DataFrame inicial
print("=== DataFrame inicial ===")
print(df)
print("\n")

```

```python
# Creamos una nueva columna: PIB per cápita = PIB / población
df["pib_per_capita"] = df["PIB"] / df["poblacion"]

# Creamos otra columna: densidad poblacional = población / área
df["densidad_poblacional"] = df["poblacion"] / df["area"]

# Mostramos el DataFrame con las columnas agregadas
print("=== Con columnas de PIB per cápita y densidad ===")
print(df)
print("\n")


```

```python
# Seleccionamos solo las columnas de PIB per cápita para los cuatro países
# (el reto pide “imprimir una tabla con los pib percapita")
df_pibpc = df[["pais", "pib_per_capita"]]

print("=== PIB per cápita por país ===")
print(df_pibpc)
print("\n")

# Seleccionamos las columnas de densidad poblacional para los cuatro países
df_densidad = df[["pais", "densidad_poblacional"]]

print("=== Densidad poblacional por país ===")
print(df_densidad)
print("\n")

```

```python
# Finalmente, guardamos el resultado final en un DataFrame llamado df_resultados
# Este DataFrame contiene país, PIB per cápita y densidad
df_resultados = df[["pais", "pib_per_capita", "densidad_poblacional"]]

print("=== df_resultados ===")
print(df_resultados)
```

```python
# Definir el nombre del archivo de salida
# No olvide cambiar en esta linea siguiente lo relacionado con su nombre justo después de IntroPython_
nombre_archivo = 'tarea_01_IntroPython_LauraVanesaNantesRoma.xlsx'

# 1. Exportar el DataFrame a Excel
# index=False es importante para no incluir la columna numérica automática de pandas.
df_resultados.to_excel(nombre_archivo, index=False)

print(f"\n¡Éxito! DataFrame exportado a '{nombre_archivo}'")
```

````python
# Para siguiente sesion mejor...
#Nivel 4: una función devuelve una tupla con varios resultados
def estadisticas(consumos):
    minimo = min(consumos)
    maximo = max(consumos)
    promedio = sum(consumos) / len(consumos)
    return minimo, maximo, promedio  # retorna una tupla

consumos = [90, 120, 110, 130]
minimo, maximo, promedio = estadisticas(consumos)

print("Mínimo:", minimo)
print("Máximo:", maximo)
print("Promedio:", promedio)
````
